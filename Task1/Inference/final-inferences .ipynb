```python
import pandas as pd
import numpy as np

def resultant_preds(path1, path2):
    weight1 = 0.1153
    weight2 = 0.1708
    df1 = pd.read_csv(path1)
    df1.index = range(len(df1))
    df2 = pd.read_csv(path2)
    df2.index = range(len(df2))
    merged_df = pd.concat([df1, df2], axis=1, join='inner')
    merged_df['likes'] = (df1 * weight1 + df2 * weight2) / (weight1 + weight2)
#     print(merged_df)
    return merged_df
# Time predictions 
T_CLIP = '/kaggle/input/clip-time/task01_submission_xgb_v2_time_CLIP.csv'
T_ENET =  '/kaggle/input/enet-time/task01_submission_xgb_v1_time_ENET.csv'
df_T = resultant_preds(T_CLIP, T_ENET)
df_T.drop(columns=['prediction', 'prediction'], inplace=True)
display(df_T)
print(df_T.describe())
df_T.to_excel(f'behaviour_simulation_test_time.xlsx',index=True)

# Company Predictions
C_CLIP = '/kaggle/input/company-clip/task01_submission_xgb_v2_Company_CLIP.csv'
C_ENET =  '/kaggle/input/company-enet/task01_submission_xgb_v1_Company_ENet.csv'
df_C = resultant_preds(C_CLIP, C_ENET)
df_C.drop(columns=['prediction', 'prediction'], inplace=True)
display(df_C)
print(df_C.describe())
df_C.to_excel(f'behaviour_simulation_test_company.xlsx',index=True)
```
